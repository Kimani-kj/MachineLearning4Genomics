```python
##loadinf a dataset from SCikit learn
from sklearn.datasets import load_iris
```

```python
##loading an algorithim for the data
from sklearn.neighbors import KNeighborsClassifier
```

```python
##loading the split that splits data 80% training and 20% testing 
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
```

```python
iris_df.head()
```

```python
iris_df['species']= iris['target']
```

```python
##visualizing the plots
from matplotlib import pyplot as  plt
iris_df.plot(kind='scatter',x='sepal length (cm)', y='petal width (cm)',
            s=32 , alpha=8, )
```

```python
x_train, x_test, y_train, y_test= train_test_split(iris.data, iris.target, random_state=6)#this makes it reproducible
knm_model=KNeighborsClassifier()
```

```python
x_train.shape
```

```python
##creating a model
knm_model.fit(x_train, y_train)
```

```python
##calculating the accuracy
print('The accuracy on the test data', round(100* knm_model.score(x_test, y_test), 2),"%")
```

```python
print('The accuracy on the training data', round(100* knm_model.score(x_train, y_train), 2),"%")
```

```python
'''## when ther performance of the training is better than on the test data the model is overfitting. 
   ##when the operdormance on the test data is beter tna the trainning data the model is 
    generalising well(as long as both performance are pretty good)
   ## when the performance of both trainig and test are poor the mode is underfitting'''
```
